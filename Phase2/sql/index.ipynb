### Subquery


```python
#example 1
complex1 = """ 
SELECT e.firstName, e.lastName, e.jobTitle
FROM employees e
JOIN offices o
     USING(officeCode)
WHERE o.country = 'France';
"""

french_df = pd.read_sql_query(complex1, conn2)

french_df.head()
```

```python
simple1 = """ 
SELECT firstName, lastName, jobTitle
FROM employees
WHERE officeCode IN (
                     SELECT officeCode
                     FROM offices o
                     WHERE o.country = 'France'   
                    )
;
"""

french_df1 = pd.read_sql_query(simple1, conn2)

french_df1.head()
```
